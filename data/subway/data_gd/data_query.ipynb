## 线路坐标数据

```python
df_line_raw = NetDataQuery(SubwayPrd).line_poses()
df_line_raw
```

```python
df_line_raw[df_line_raw['city_name'] == '上海市']
```

```python
# 线路id
df_line_id = df_line_raw[['line_id', 'line_name', 'city_id']]
df_line_id = df_line_id.drop_duplicates(subset=['city_id', 'line_name'], keep='first').reset_index(drop=True)
# 添加线路line_order列，表示线路的顺序
# df_line_id['line_order'] = df_line_id.groupby('city_id').cumcount() + 1
df_line_id
```

```python
df_line = df_line_raw.copy()
# city_name简写
df_line['city_name'] = df_line['city_name'].apply(lambda x: x.replace('市', ''))
df_line['city_name'] = df_line['city_name'].apply(lambda x: x.replace('特别行政区', ''))
# 将line_poses转为列表
df_line['line_poses_list'] = df_line['line_poses'].apply(lambda x: [i for i in x.split(',')])
# 添加线路line_order列，表示线路的顺序
df_line['line_order'] = df_line.groupby('city_id').cumcount() + 1
# st_count列，即每条线路的站点数量
df_line['st_count'] = df_line['line_poses_list'].apply(len)
df_line
```

```python
df_line[df_line['city_name'] == '上海']
```

```python
# 将line_poses_list拆分为多行
df_line = df_line.explode('line_poses_list').reset_index(drop=True)
# 将line_poses_list拆分为pos_x, pos_y两列
df_line[['pos_x', 'pos_y']] = df_line['line_poses_list'].str.split(' ', expand=True)
df_line
```

```python
# 按city_id, line_name分组，增加st_order列, 即为每条线路的每个站点的顺序
df_line['line_st_order'] = df_line.groupby(['city_id', 'line_id']).cumcount() + 1
# 添加line_st_id列，即为每条线路的每个站点的唯一标识
df_line['st_id_line'] = df_line['line_st_order']
df_line['target_st_id_line'] = df_line['st_id_line'] + 1
df_line['target_st_id_line'] = df_line['target_st_id_line'].astype('Int64')
# 删除target_st_id中大于st_count的值，避免出现不存在的站点
df_line.loc[df_line['target_st_id_line'] > df_line['st_count'], 'target_st_id_line'] = None
# 添加整体st_id列，唯一值，同一站点存在多个线路时，st_id不同
df_line['st_id'] = df_line['city_id'].astype(str) +"c" +  df_line['line_order'].astype(str).str.zfill(2) + "l" + df_line['st_id_line'].astype(str).str.zfill(3)
# 添加整体target_st_id列，如果target_st_id_line为空，则为空，否则为city_id + line_order + target_st_id_line，唯一值
df_line['target_st_id'] = df_line['city_id'].astype(str) +"c" +  df_line['line_order'].astype(str).str.zfill(2) + "l" + df_line['target_st_id_line'].astype(str).str.zfill(3)
df_line.loc[df_line['target_st_id_line'].isna(), 'target_st_id'] = None
# 添加st_poiid列，即同一站点的唯一标识, 按照坐标值识别同一个站点
df_line['st_poiid'] = "c" + df_line['city_id'].astype(str)  +  'x' + df_line['pos_x'].astype(str) + 'y' + df_line['pos_y'].astype(str)

df_line
```

```python
df_st_id = df_line[['st_id', 'st_poiid']]
df_line = df_line.merge(df_st_id, left_on='target_st_id', right_on='st_id', how='left')
df_line = df_line.drop(columns=['st_id_y'])
df_line = df_line.rename(columns={'st_id_x': 'st_id','st_poiid_x': 'st_poiid',  'st_poiid_y': 'target_st_poiid'})
df_line
```

```python
# 统一line_id列
df_line = df_line.drop(columns=['line_id'])
df_line = df_line.merge(df_line_id, on=['city_id', 'line_name'], how='left')
df_line
```

```python
# 删除同线路名的支线数据中的重复站点数据
df_line = df_line.drop_duplicates(subset=['city_id', 'line_name', 'st_poiid', 'target_st_poiid'], keep='first').reset_index(drop=True)
df_line
```

```python
# 添加空列 'll_distance', 'longitude', 'latitude','longitude_bd','latitude_bd'， 值都设为1
df_line['ll_distance'] = 1
df_line['longitude'] = 1
df_line['latitude'] = 1
df_line['longitude_bd'] = 1
df_line['latitude_bd'] = 1
df_line['st_name'] = df_line['st_poiid']
df_line['target_st_name'] = df_line['target_st_poiid']

# 保存为csv
df_line.to_csv("ads_subway_line_pos.csv", index=False)
```
